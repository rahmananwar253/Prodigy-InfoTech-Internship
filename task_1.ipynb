### Create a bar chart or histogram to visualize the distribution of a categorical or continuous variable, such as the distribution of ages or genders in a population.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('sub-division_population_of_pakistan.csv')     
df.head()
```

```python
# column names
df.columns
```

```python
# combine male(rural) and female(rural) column into a new column named gender
df['gender'] = df['MALE (RURAL)'] + df['FEMALE (RURAL)']
```

### Plot using seaborn

```python
# plotting gender distribution across provinces using barplot
sns.barplot(data=df, x='PROVINCE',y='gender',hue='PROVINCE')
```

```python
# plotting gender distribution across provinces using histplot
sns.histplot(df, x="gender", hue="PROVINCE")
```

```python
# plotting transgender(urban) distribution across provinces using barplot
sns.barplot(data=df, x='PROVINCE',y='TRANSGENDER (URBAN)',hue='PROVINCE') 
```

```python
# plotting transgender(rural) distribution across provinces using histplot
sns.histplot(df, x="TRANSGENDER (RURAL)", hue="PROVINCE")
```

```python
# plotting area distribution across provinces using barplot
sns.barplot(data=df, x='PROVINCE',y='AREA (sq.km)',hue='PROVINCE')
```

```python
# plotting annual growth rate(rural) distribution across provinces using barplot
sns.barplot(data=df, x='PROVINCE',y='ANNUAL GROWTH RATE (RURAL)',hue='PROVINCE')
plt.title('Annual Growth Rate (Rural) Distribution Across Provinces')
```

```python
# plotting gender distribution across provinces using histogram
df['gender'].plot(kind='hist', bins=5)
```

```python
# plotting gender distribution across provinces using barplot
df.groupby('PROVINCE')['gender'].sum().plot(kind='bar')
plt.title('Gender Distribution across Provinces')
plt.xlabel('Provinces')
plt.ylabel('Gender')
plt.xticks(rotation=45)
plt.show()
```

```python
# plotting transgender(rural) distribution across provinces using histogram
df['TRANSGENDER (RURAL)'].plot(kind='hist', bins=5)
plt.title('Transgender Distribution')
plt.xticks(rotation=45)
plt.show()
```

```python
# plotting transgender(rural) distribution across provinces using barplot
ax=df.groupby('PROVINCE')['TRANSGENDER (RURAL)'].sum().plot(kind='bar')
plt.title('Transgender Distribution across Provinces')
plt.xlabel('Provinces')
plt.ylabel('Transgender')
plt.xticks(rotation=45)
# add text on the bars
for p in ax.patches:
    ax.text(p.get_x() + p.get_width() / 2., p.get_height(),
            f'{p.get_height():.2f}',  # Format the height value
            ha='center', va='bottom')  # Center-align horizontally and bottom-align vertically

plt.show()
```

```python
color=[ 'green', 'yellow', 'orange', 'cyan', 'magenta']
ax=df.groupby('PROVINCE')['AVG HOUSEHOLD SIZE (RURAL)'].sum().plot(kind='bar', color=color)
plt.title('Average Household Size (Rural) across Provinces')
plt.xlabel('Provinces')
plt.ylabel('Average Household Size (Rural)')
plt.xticks(rotation=45)
# add text on the bars
for p in ax.patches:
    ax.text(p.get_x() + p.get_width() / 2., p.get_height(),
            f'{p.get_height():.2f}',  # Format the height value
            ha='center', va='bottom')  # Center-align horizontally and bottom-align vertically

plt.show()
```
